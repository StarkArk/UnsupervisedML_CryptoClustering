```python
# Import required libraries and dependencies
import pandas as pd
import matplotlib.pyplot as plt
import hvplot.pandas
import os
os.environ['OMP_NUM_THREADS'] = '1'
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data 
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Display Info and DataFrame as reference
display(df_market_data.info(),df_market_data.head(5))
```

```python
# Normalize the data from the CSV file
crypto_market_scaled = StandardScaler().fit_transform(df_market_data)

# First ten rows of the scaled data
crypto_market_scaled[0:10]
```

```python
df_market_data.columns
```

```python
# Create a DataFrame with the scaled data
df_crypto_market_scaled = pd.DataFrame(crypto_market_scaled, columns=['price_change_percentage_24h', 'price_change_percentage_7d',
       'price_change_percentage_14d', 'price_change_percentage_30d', 'price_change_percentage_60d', 'price_change_percentage_200d',
       'price_change_percentage_1y'])

# Copy the crypto names from the original data and add to df
df_crypto_market_scaled['coin_id'] = list(df_market_data.index)

# Set coinid as index
df_crypto_market_scaled = df_crypto_market_scaled.set_index('coin_id')

# Display sample data
df_crypto_market_scaled.head(10)
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1,12))
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
for i in k:
    k_model = KMeans(n_clusters=i, random_state=0)
    k_model.fit(df_crypto_market_scaled)
    inertia.append(k_model.inertia_)
```

```python
# Create a DataFrame with the data to plot the Elbow curve
elbow_df = pd.DataFrame({"k": k, "inertia": inertia})

elbow_df.head()
```

```python
# Plot a line chart with all the inertia values
elbow_df.hvplot.line(
    x = 'k',
    y = 'inertia',
    xticks = k,
    width = 800,
    height = 500

)
```

#### Answer the following question: 

What is the best value for `k`?

**4**

---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialize the K-Means model using k = 4
model = KMeans(n_clusters=4) 
```

```python
# Fit the K-Means model using the scaled data
model.fit(df_crypto_market_scaled)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
kmeans_predictions = model.predict(df_crypto_market_scaled)

# Print the resulting array of cluster values.
kmeans_predictions
```

```python
# Create a copy of the DataFrame
crypto_clustered_df = df_crypto_market_scaled.copy()
```

```python
# Add a new column to the DataFrame with the predicted clusters
crypto_clustered_df['kmeans_cluster'] = kmeans_predictions

# Display sample data
crypto_clustered_df.sample(5)
```

```python
# Create scatter plot of clusters using 'price_change_percentage_24h' and 'price_change_percentage_7d'
crypto_clustered_df.hvplot.scatter(
    x="price_change_percentage_24h", 
    y="price_change_percentage_7d", 
    by="kmeans_cluster",
    hover_cols='coin_id',
    width=800,
    height=500,
    size=50
)
```

---

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance setting `n_components=3`.
pca=PCA(n_components=3)
```

```python
# Use the PCA model to reduce to three principal components.
crypto_pca = pca.fit_transform(df_crypto_market_scaled)
# View the first five rows of the DataFrame. 
crypto_pca[0:5]
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.
pca_attribution = sum(pca.explained_variance_ratio_)
pca_att_list = list(pca.explained_variance_ratio_)

# display(sum(pca.explained_variance_ratio_), list(pca.explained_variance_ratio_))
print(f'PCA Explained Variance: {round(100 * pca_attribution, 2)}%')
print(f'First Component: {round(100 * pca_att_list[0], 2)}%')
print(f'Second Component: {round(100 * pca_att_list[1], 2)}%')
print(f'Third Component: {round(100 * pca_att_list[2], 2)}%')
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**89.5%**

```python
# Create a new DataFrame with the PCA data.
crypto_pca_df = pd.DataFrame(crypto_pca, columns=['PC1', 'PC2', "PC3"])
# Creating a DataFrame with the PCA data

# Copy the crypto names from the original data
crypto_pca_df['coin_id'] = list(df_market_data.index)

# Set the coinid column as index
crypto_pca_df = crypto_pca_df.set_index('coin_id')

# Display sample data
crypto_pca_df.head(10)
```

---

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1,12))
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
for i in k:
    k_model = KMeans(n_clusters=i, random_state=0)
    k_model.fit(crypto_pca_df)
    inertia.append(k_model.inertia_)
```

```python
# Create a DataFrame with the data to plot the Elbow curve
pca_elbow_df = pd.DataFrame({"k": k, "inertia": inertia})

pca_elbow_df.head()
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
pca_elbow_df.hvplot.line(
    x = 'k',
    y = 'inertia',
    xticks = k,
    width = 800,
    height = 500

)
```

#### Answer the following questions: 

* What is the best value for `k` when using the PCA data?

  * **4**


* Does it differ from the best k value found using the original data?

  * **No**

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialize the K-Means model using the best value for k
model = KMeans(n_clusters=4) 
```

```python
# Fit the K-Means model using the PCA data
model.fit(crypto_pca_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the PCA data
pca_crypto_model = model.predict(crypto_pca_df)
# Print the resulting array of cluster values.
pca_crypto_model
```

```python
# Create a copy of the DataFrame with the PCA data
crypto_pca_cluster_df = crypto_pca_df.copy()

# Add a new column to the DataFrame with the predicted clusters
crypto_pca_cluster_df['pca_cluster'] = pca_crypto_model

# Display sample data
crypto_pca_cluster_df.head(10)
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="PC1"` and `y="PC2"`
crypto_pca_cluster_df.hvplot.scatter(
    x="PC1", 
    y="PC2", 
    by="pca_cluster",
    hover_cols='coin_id',
    width=800,
    height=500,
    size=50
)
```

### Visualize and Compare the Results

In this section, you will visually analyze the cluster analysis results by contrasting the outcome with and without using the optimization techniques.

```python
# Composite plot to contrast the Elbow curves
elbow_df['pca_inertia'] = pca_elbow_df['inertia']

elbow_df.hvplot.line(x = 'k', y = ['inertia', 'pca_inertia'], xticks = k, width = 800, height = 500, color= ['darkorange', 'midnightblue'], 
                     title='Elbow Diagram of Full Featured Clustering vs. 3 component PCA')

# plt.savefig('Images/comp_elbow_plot.png')
```

```python
# Composite plot to contrast the clusters
crypto_clustered_df['pca_cluster'] = crypto_pca_cluster_df['pca_cluster']

crypto_scatter_plot = (
    crypto_clustered_df.hvplot.scatter(x="price_change_percentage_24h", y="price_change_percentage_7d", subplots=True, by="kmeans_cluster",hover_cols='coin_id',width=150,height=200, title='KMeans') +
    crypto_clustered_df.hvplot.scatter(x="price_change_percentage_24h", y="price_change_percentage_7d", subplots=True, by="pca_cluster",hover_cols='coin_id',width=150,height=200, title='PCA')
    )

crypto_scatter_plot2 = (
    crypto_clustered_df.hvplot.scatter(x="price_change_percentage_24h", y="price_change_percentage_60d", subplots=True, by="kmeans_cluster",hover_cols='coin_id',width=150,height=200, title='KMeans') +
    crypto_clustered_df.hvplot.scatter(x="price_change_percentage_24h", y="price_change_percentage_60d", subplots=True, by="pca_cluster",hover_cols='coin_id',width=150,height=200, title='PCA')
    )

crypto_scatter_plot3 = (
    crypto_clustered_df.hvplot.scatter(x="price_change_percentage_24h", y="price_change_percentage_1y", subplots=True, by="kmeans_cluster",hover_cols='coin_id',width=150,height=200, title='KMeans') +
    crypto_clustered_df.hvplot.scatter(x="price_change_percentage_24h", y="price_change_percentage_1y", subplots=True, by="pca_cluster",hover_cols='coin_id',width=150,height=200, title='PCA')
    )

# Look at the 24h change vs. the 7d, 60d, and 1y to compare clusters
display(crypto_scatter_plot, crypto_scatter_plot2, crypto_scatter_plot3)
```
